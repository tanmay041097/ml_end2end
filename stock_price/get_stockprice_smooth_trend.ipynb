```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib import style
```

```python
df_google = pd.read_csv("GOOG.csv")
df_google.head()
```

```python
df_google.info() # check the data types in google csv dataframe 
```

```python
df_google.dtypes # short - check the data types in google csv dataframe
```

```python
df_google['date'] = pd.to_datetime(df_google['date'], format='%Y-%m-%d')
df_google.info()
```

```python
style.use("ggplot")
df_google.plot(x="date", y="close", figsize=(16,8))
plt.show()
```

```python
column_4 =  df_google["close"].to_numpy()  
print(column_4)

avg_stock_price = []
i = 0
window_size = 35          
n = len(column_4)
print(n)

while i < n - window_size - 1:                              #Average of elements upto windows size
    average = round(np.sum(column_4[i:i+window_size]/window_size),2)
    avg_stock_price.append(average)
    i+=1

print(avg_stock_price)

no_noise = { "close": avg_stock_price}                     # Converting numpy array to Pandas Data series
data_with_no_noise = pd.DataFrame(no_noise)


data_with_no_noise["data"] = df_google["date"]
style.use("ggplot")                                         #Plot
data_with_no_noise.plot(x="data", y="close", figsize=(16,8))
plt.show()
```
